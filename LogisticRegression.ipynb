# LinearRegression

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
N = 1000
left = -20
right = 60

X = np.random.uniform(left, right, N)
error = (np.random.random(N) - 0.5)*4.0
y = 0.25*X + 5.0 + error
```

```python
plt.scatter(X, y)
```

```python
X = X.reshape(-1,1)
```

```python
# y = k*x + b
```

```python
lr = LinearRegression() # s1: create instance
lr.fit(X, y) # s2: fit
y_pred = lr.predict(X) # s3: predict
```

```python
lr.coef_, lr.intercept_
```

```python
plt.scatter(X, y)
plt.scatter(X, y_pred)
```
